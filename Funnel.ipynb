```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dtypes = {
    'visitorid': 'int32',
    'event': 'category',
    'itemid': 'int32',
    'transactionid': 'Int64'   # nullable, mostly NaN
}
path = r'C:\Users\Temo\OneDrive\Desktop\Product Data\events.csv' 

events = pd.read_csv(
    path,
    dtype=dtypes,
    low_memory=False
)

events['event_time'] = pd.to_datetime(events['timestamp'], unit='ms', utc=True)
```

```python
events['visitorid'].isna().sum()
```

```python
events = events.sort_values(['visitorid','event_time'])
```

```python
events = events[['visitorid', 'itemid', 'event', 'event_time']]
```

```python
events['event'].value_counts()
```

```python
# First view time
first_view = (events[events['event'] == 'view']
              .groupby('visitorid')['event_time'].min()
              .rename('first_view_time'))

# First add-to-cart  
first_cart = (events[events['event'] == 'addtocart']
              .groupby('visitorid')['event_time'].min()
              .rename('first_cart_time'))

# First transaction
first_transaction = (events[events['event'] == 'transaction']
                     .groupby('visitorid')['event_time'].min()
                     .rename('first_transaction_time'))

# Join all together
visitor_dates = (pd.concat([first_view, first_cart, first_transaction], axis=1)
                .fillna(pd.NaT))  # NaT for visitors without that event
```
