```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dtypes = {
    'visitorid': 'int32',
    'event': 'category',
    'itemid': 'int32',
    'transactionid': 'Int64'   # nullable, mostly NaN
}
path = r'data\events.csv' 

events = pd.read_csv(
    path,
    dtype=dtypes,
    low_memory=False
)

events['event_time'] = pd.to_datetime(events['timestamp'], unit='ms', utc=True)
```

```python
events = events.sort_values(['visitorid','event_time'])
```

```python
events = events[['visitorid', 'itemid', 'event', 'event_time']]
```

```python
events['event'].value_counts()
```

```python
# First view time
first_view = (events[events['event'] == 'view']
              .groupby('visitorid')['event_time'].min()
              .rename('first_view_time'))

# First add-to-cart  
first_cart = (events[events['event'] == 'addtocart']
              .groupby('visitorid')['event_time'].min()
              .rename('first_cart_time'))

# First transaction
first_transaction = (events[events['event'] == 'transaction']
                     .groupby('visitorid')['event_time'].min()
                     .rename('first_transaction_time'))

# Join all together
visitor_dates = (pd.concat([first_view, first_cart, first_transaction], axis=1)
                .fillna(pd.NaT))  # NaT for visitors without that event
```

```python
funnel_df = visitor_dates[visitor_dates['first_view_time'].notna()]
```

```python
mask_added_7d = (
    (funnel_df['first_cart_time'] >= funnel_df['first_view_time']) &
    (funnel_df['first_cart_time'] <= funnel_df['first_view_time'] + pd.Timedelta(days=7))
)
funnel_df['added_7d'] = mask_added_7d.astype(int)
```

```python
mask_purchase_7d = (
    (funnel_df['first_transaction_time'] >= funnel_df['first_view_time']) &
    (funnel_df['first_transaction_time'] <= funnel_df['first_view_time'] + pd.Timedelta(days=7))
)

funnel_df['purchased_7d'] = mask_purchase_7d.astype(int)
```

```python
funnel_df['added_30d'] = (
    (funnel_df['first_cart_time'] >= funnel_df['first_view_time']) &
    (funnel_df['first_cart_time'] <= funnel_df['first_view_time'] + pd.Timedelta(days=30))
).astype(int)

funnel_df['purchased_30d'] = (
    (funnel_df['first_transaction_time'] >= funnel_df['first_view_time']) &
    (funnel_df['first_transaction_time'] <= funnel_df['first_view_time'] + pd.Timedelta(days=30))
).astype(int)
```

```python
funnel_df = funnel_df[['first_view_time', 'added_7d', 'purchased_7d', 'added_30d', 'purchased_30d']]
```

```python
funnel_df.rename(columns={'first_view_time': 'entry_time'}, inplace=True)
```

```python
views = len(funnel_df)
adds = funnel_df['added_7d'].sum()
purchases = funnel_df['purchased_7d'].sum()
adds_30 = funnel_df['added_30d'].sum()
purchases_30 = funnel_df['purchased_30d'].sum()
```

```python
view_to_add = adds / views
add_to_purchase_7 = purchases / adds
view_to_purchase_7 = purchases / views
add_to_purchase_30 = purchases_30 / adds_30
view_to_purchase_30 = purchases_30 / views
```

```python
print(view_to_add)
print(add_to_purchase_7)
print(view_to_purchase_7)
print(add_to_purchase_30)
print(view_to_purchase_30)
```
